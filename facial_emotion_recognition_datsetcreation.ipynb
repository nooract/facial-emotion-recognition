```python


from PIL import Image
import os
# Path to the directory containing the images
input_dir =  "/content/frames"
# Path to the directory where the resized images will be saved
output_dir = "/content/gdrive/MyDrive/frames_resized"
# Target size of the resized images
target_size = (224, 224)

# Loop through all the files and directories in the input directory
for root, dirs, files in os.walk(input_dir):
    for file in files:
        # Check if the file is an image
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            # Construct the path to the input image file
            input_path = os.path.join(root, file)
            # Construct the path to the output image file
            output_path = os.path.join(output_dir, os.path.relpath(input_path, input_dir))
            # Create the directory for the output image file if it doesn't exist
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            # Open the image using the Pillow library
            image = Image.open(input_path)
            # Resize the image to the target size
            resized_image = image.resize(target_size)
            # Save the resized image to the output directory with the same filename
            resized_image.save(output_path)
```
